```python
# Import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
max = 100
total = [*range(1,max)]
total
inputurl = "https://www.idnfinancials.com/company/industry/food-and-beverages-42"

# Create an URL object
for pagenum in range(1,max):
  url = inputurl + '/' + str(pagenum)
  code = url.split('/')[5].split('-')[0]
  # Create object page
  page = requests.get(url)
  soup = BeautifulSoup(page.text, 'lxml')
  # Obtain information from tag <table>
  table1 = soup.find('div', id="table-companies")
  # Obtain every title of columns with tag <th>
  headers = []
  url_list = []
  try:
    for i in table1.find_all('a', class_ = "table-sorter"):
      title = i.text
      headers.append(title)
    # Create a dataframe
    mydata1 = pd.DataFrame(columns = headers)
    # Create a for loop to fill mydata1
    for j in table1.find_all(class_ = "table-row")[1:]:
      row_data = j.find_all("div")
      row = [i.text for i in row_data]
      length = len(mydata1)
      mydata1.loc[length] = row
    table1 = soup.find('div', id="table-companies")
    for y in table1.find_all(class_ = "table-row")[1:]:
      for a in y.find_all('a', href=True):
        url_list.append(a['href'])
    
    mydata1['URL'] = url_list

    mydata1['Company']=mydata1['Company'].str[1:5]
    # Export to csv
    namecsv = 'companylist'+ str(pagenum) + '.csv'
    mydata1.to_csv(namecsv, index=False)
  except:
    print("fail")
    total.remove(pagenum)
max_merge = total[-1]
merge_csv_list = list()

newdf_append = pd.DataFrame()

for z in range(1,max_merge+1):
  merge_name_csv = 'companylist'+ str(z) + '.csv'
  merge_csv_list.append(merge_name_csv)

sector = code

df_csv_concat_all = pd.concat([pd.read_csv(file) for file in merge_csv_list], ignore_index=True)
df_csv_concat_all.to_csv("data-"+ sector +".csv")
```

```python
counter = -1
for dd in df_csv_concat_all['URL']:
  counter = counter + 1
  max = 15
  total = [*range(1,max)]
  total
  
  try:
    print(str(dd)+"/dividend/")

    inputurl = str(dd)+"/dividend/"
    
    # Create an URL object
    for pagenum in range(1,max):
      url = inputurl + str(pagenum)
      code = url.split('/')[3]
      # Create object page
      page = requests.get(url)
      soup = BeautifulSoup(page.text, 'lxml')
      # Obtain information from tag <table>
      table1 = soup.find('table', id='table-dividend')
      # Obtain every title of columns with tag <th>
      try:
        headers = []
        for i in table1.find_all('th'):
          title = i.text
          headers.append(title)
        # Create a dataframe
        mydata3 = pd.DataFrame(columns = headers)
        # Create a for loop to fill mydata3
        for j in table1.find_all('tr')[1:]:
          row_data = j.find_all('td')
          row = [i.text for i in row_data]
          length = len(mydata3)
          mydata3.loc[length] = row
        # Drop and clearing unnecessary rows
        mydata3.drop('Bonus Shares', inplace=True, axis=1)
        # Drop and clearing unnecessary rows
        mydata3['Cash Dividend']=mydata3['Cash Dividend'].map(lambda x: x.rstrip('IDRUSD()'))
        mydata3['Cash Dividend']=mydata3['Cash Dividend'].astype(float)
        mydata3['Year']=mydata3['Year'].astype(int)
        # Export to csv
        namecsv = 'pure-dividend-'+ str(code)+ "-"+ str(pagenum) + '.csv'
        mydata3.to_csv(namecsv, index=False)
      except:
        total.remove(pagenum)

    max_merge = total[-1]
    merge_csv_list = list()

    newdf_append = pd.DataFrame()

    for z in range(1,max_merge+1):
      merge_name_csv = 'pure-dividend-'+ str(code)+ "-"+ str(z) + '.csv'
      merge_csv_list.append(merge_name_csv)

    df_csv_concat = pd.concat([pd.read_csv(file) for file in merge_csv_list], ignore_index=True)

    groupdatatotal = df_csv_concat.groupby('Year').sum()
    groupdatatotal = groupdatatotal.sort_values(by=['Year'], ascending = False)

    df_finalonly = df_csv_concat
    df_finalonly = df_finalonly[df_finalonly.Type == 'Final']
    df_finalonly_sum = df_finalonly.groupby('Year').sum()
    df_finalonly_sum = df_finalonly_sum.sort_values(by=['Year'], ascending = False)
    df_finalonly_sum.T
    df_finalonly_sum.to_csv("finalonly-total-data-"+ code +".csv")
  except:
    headers1 = ['Year', 'Cash Dividend']
    df_finalonly_sum = pd.DataFrame(columns = headers1)
    df_finalonly_sum.loc[0] = [0, 0]
    df_finalonly_sum = df_finalonly_sum.groupby('Year').sum()
    df_finalonly_sum = df_finalonly_sum.sort_values(by=['Year'], ascending = False)
    df_finalonly_sum.T
    df_finalonly_sum.to_csv("finalonly-total-data-"+ code +".csv")
    pass
```
